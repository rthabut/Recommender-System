On observe qu'une bonne portion des films appartient donc à plusieurs genres.
On va donc essayer de regrouper les genres les plus proches entres eux. Pour ce faire on va utiliser la matrice des corrélation entre genres. 
On l'utilise comme matrice de distance, pour, ensuite, avec un algorithme de hierachical clustering ascendant, déterminer quels genres sont les plus similaires.




```python
def heatmap(df):
    df_new=df.copy()
    del df_new['MovieID'] #On ne veut pas faire de régression sur l'ID du film 
    corr_matrix = df_new.corr()
    mask = np.zeros_like(corr_matrix)
    mask[np.triu_indices_from(mask)] = True
    with sns.axes_style("white"):
        plt.subplots(figsize=(20,14))
        plt.title('Corrélations des catégories et clusters de films entre eux')
        sns.heatmap(corr_matrix, mask=mask, annot=True);
```

## Une première ACP

```python
def display_circles(pcs, n_comp, pca, axis_ranks, labels=None, label_rotation=0, lims=None):
    for d1, d2 in axis_ranks: # On affiche les 3 premiers plans factoriels, donc les 6 premières composantes
        if d2 < n_comp:

            # initialisation de la figure
            fig, ax = plt.subplots(figsize=(7,6))

            # détermination des limites du graphique
            if lims is not None :
                xmin, xmax, ymin, ymax = lims
            elif pcs.shape[1] < 30 :
                xmin, xmax, ymin, ymax = -1, 1, -1, 1
            else :
                xmin, xmax, ymin, ymax = min(pcs[d1,:]), max(pcs[d1,:]), min(pcs[d2,:]), max(pcs[d2,:])

            # affichage des flèches
            # s'il y a plus de 30 flèches, on n'affiche pas le triangle à leur extrémité
            if pcs.shape[1] < 30 :
                plt.quiver(np.zeros(pcs.shape[1]), np.zeros(pcs.shape[1]),
                   pcs[d1,:], pcs[d2,:], 
                   angles='xy', scale_units='xy', scale=1, color="grey")
                # (voir la doc : https://matplotlib.org/api/_as_gen/matplotlib.pyplot.quiver.html)
            else:
                lines = [[[0,0],[x,y]] for x,y in pcs[[d1,d2]].T]
                ax.add_collection(LineCollection(lines, axes=ax, alpha=.1, color='black'))
            
            # affichage des noms des variables  
            if labels is not None:  
                for i,(x, y) in enumerate(pcs[[d1,d2]].T):
                    if x >= xmin and x <= xmax and y >= ymin and y <= ymax :
                        plt.text(x, y, labels[i], fontsize='14', ha='center', va='center', rotation=label_rotation, color="blue", alpha=0.5)
            
            # affichage du cercle
            circle = plt.Circle((0,0), 1, facecolor='none', edgecolor='b')
            plt.gca().add_artist(circle)

            # définition des limites du graphique
            plt.xlim(xmin, xmax)
            plt.ylim(ymin, ymax)
        
            # affichage des lignes horizontales et verticales
            plt.plot([-1, 1], [0, 0], color='grey', ls='--')
            plt.plot([0, 0], [-1, 1], color='grey', ls='--')

            # nom des axes, avec le pourcentage d'inertie expliqué
            plt.xlabel('F{} ({}%)'.format(d1+1, round(100*pca.explained_variance_ratio_[d1],1)))
            plt.ylabel('F{} ({}%)'.format(d2+1, round(100*pca.explained_variance_ratio_[d2],1)))

            plt.title("Cercle des corrélations (F{} et F{})".format(d1+1, d2+1))
            plt.show(block=False)
        
def display_factorial_planes(X_projected, n_comp, pca, axis_ranks, labels=None, alpha=1, illustrative_var=None):
    for d1,d2 in axis_ranks:
        if d2 < n_comp:
 
            # initialisation de la figure       
            fig = plt.figure(figsize=(7,6))
        
            # affichage des points
            if illustrative_var is None:
                plt.scatter(X_projected[:, d1], X_projected[:, d2], alpha=alpha)
            else:
                illustrative_var = np.array(illustrative_var)
                for value in np.unique(illustrative_var):
                    selected = np.where(illustrative_var == value)
                    plt.scatter(X_projected[selected, d1], X_projected[selected, d2], alpha=alpha, label=value)
                plt.legend()

            # affichage des labels des points
            if labels is not None:
                for i,(x,y) in enumerate(X_projected[:,[d1,d2]]):
                    plt.text(x, y, labels[i],
                              fontsize='14', ha='center',va='center') 
                
            # détermination des limites du graphique
            boundary = np.max(np.abs(X_projected[:, [d1,d2]])) * 1.1
            plt.xlim([-boundary,boundary])
            plt.ylim([-boundary,boundary])
        
            # affichage des lignes horizontales et verticales
            plt.plot([-100, 100], [0, 0], color='grey', ls='--')
            plt.plot([0, 0], [-100, 100], color='grey', ls='--')

            # nom des axes, avec le pourcentage d'inertie expliqué
            plt.xlabel('F{} ({}%)'.format(d1+1, round(100*pca.explained_variance_ratio_[d1],1)))
            plt.ylabel('F{} ({}%)'.format(d2+1, round(100*pca.explained_variance_ratio_[d2],1)))

            plt.title("Projection des individus (sur F{} et F{})".format(d1+1, d2+1))
            plt.show(block=False)

def display_scree_plot(pca):
    scree = pca.explained_variance_ratio_*100
    plt.bar(np.arange(len(scree))+1, scree)
    plt.plot(np.arange(len(scree))+1, scree.cumsum(),c="red",marker='o')
    plt.xlabel("rang de l'axe d'inertie")
    plt.ylabel("pourcentage d'inertie")
    plt.title("Eboulis des valeurs propres")
    plt.show(block=False)
```

```python
# On va maintenant s'intéresser à l'influence du sexe et de l'âge sur la notation:
joint_aux=ratings.merge(movies, on = 'MovieID', how = 'inner')
joint=joint_aux.merge(users, left_on='UserID',right_on='UserID')
del joint['rating']
del joint['Timestamp']

Genre_regroup=[['Action','Adventure','Sci-Fi','Fantasy'],['Documentary','War','Western'],['Children\'s','Animation','Musical'],['Crime','Film-Noir','Thriller','Mystery'],['Comedy','Romance'],['Drama'],['Horror']]

#Objectif: parcourir les films (évite les Nan), à chaque film,repérer l'utilisateur, le genre, et lui attribuer le + rating dans l'index de la liste attribué au genre, diviser à la fin par le nombre de films notés dans le genre (ajouter un à son diviseur)

#Pour Femmes ou pour Hommes:
indiv_pref=pd.DataFrame()
for i in range(len(Genre_regroup)):
    df1=pd.DataFrame()
    for j in range(len(Genre_regroup[i])):
        index=joint['Genres'].str.contains(Genre_regroup[i][j])
        df8=joint[index].pivot(index='UserID',columns='MovieID',values='Rating')
        df1=pd.concat([df1,df8], axis = 1)
    df1_mean=df1.mean(axis=1)
    indiv_pref=pd.concat([indiv_pref,df1_mean],axis=1)
       
Colonnes=['Action,Adventure,Sci-Fi,Fantasy','Documentary,War,Western','Children\'s, Animation,Musical','Crime,Film-Noir,Thriller,Mystery','Comedy,Romance','Drama','Horror']   

indiv_pref.columns=Colonnes
indiv_pref_user=indiv_pref.merge(users,right_on='UserID',left_index=True)
del indiv_pref_user['Occupation']
del indiv_pref_user['Zip-code']
index_m=indiv_pref_user['Gender'].str.contains('M')
pref_m=indiv_pref_user[index_m]
del pref_m['Age']
del pref_m['UserID']
index_f=indiv_pref_user['Gender'].str.contains('F')
pref_f=indiv_pref_user[index_f]
del pref_f['Age']
del pref_f['UserID']
pref_m_mean = pd.DataFrame(pref_m.mean())
pref_f_mean = pd.DataFrame(pref_f.mean())
pref_sexe=pd.concat([pref_f_mean,pref_m_mean],axis=1)
pref_sexe.columns = ['femmes', 'hommes']

index_age=indiv_pref_user['Age']
index_jeune=pd.isna(index_age.mask(index_age<36))
pref_jeune=indiv_pref_user[index_jeune]
index_agé=pd.isna(index_age.mask(index_age>35))
pref_agé=indiv_pref_user[index_agé]
del pref_agé['Gender']
del pref_agé['UserID']
del pref_agé['Age']
del pref_jeune['Gender']
del pref_jeune['UserID']
del pref_jeune['Age']
pref_jeune_mean=pd.DataFrame(pref_jeune.mean())
pref_agé_mean=pd.DataFrame(pref_agé.mean())

pref_age=pd.concat([pref_jeune_mean,pref_agé_mean],axis=1)

pref_age.columns=['jeune','âgé']

pref_age.plot.bar(rot=45, title="Note moyenne attribuée à chaque groupe de genres selon l'âge");
pref_sexe.plot.bar(rot=45, title="Note moyenne attribuée à chaque groupe de genres selon le sexe");
plt.show(block=True);
Ecart_sexe=pd.DataFrame((pref_f_mean-pref_m_mean)*20)
Ecart_age=pd.DataFrame((pref_jeune_mean-pref_agé_mean)*20)
Ecart_sexe.columns=['Ecart en pourcentage entre les notes pour chaque sexe']
Ecart_age.columns=['Ecart en pourcentage entre les notes selon l\'age']
Ecart_sexe
```
